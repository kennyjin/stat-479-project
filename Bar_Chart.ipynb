```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
objects = ('CNN(ResNet-34)', 'CNN(LeNet-5)', 'k-NN', 'Random Forest', 'Logistic Regression')
y_pos = np.arange(len(objects))
accuracy = [59.96,46.79,33.05,26.86,23.96]

plt.figure(figsize=(10, 7))  # width:20, height:3
plt.bar(y_pos, accuracy, align='center', alpha=0.5)
plt.xticks(y_pos, objects, fontsize=11)
plt.ylabel('Accuracy (%)', {'fontsize': 15})
plt.title('Prediction Accuracy on the Test Set', {'fontsize': 20})

for i, v in enumerate(accuracy):
    plt.text(y_pos[i] - 0.2, v + 0.7, str(v) + "%")

plt.show()
```
